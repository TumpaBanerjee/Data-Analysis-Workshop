# What is Seaborn? 
Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. 
You will find official documentation of seaborn at https://seaborn.pydata.org/

# Advantages over Matplotlib for statistical visualization
*  High-Level API: can easily create a regression line, confidence interval etc
*  Built-in datasets and themes: it comes with a set of datasets and themes that can be explored for practicing data visualization
*  Statistical Report: can visualize the results of statistical computation, such as distribution, correlation and regression
*  Categorical plots: it supports a rich set of categorical plots
*  Facet grids and rel plots: allow to create multiple plots and charts in a single figure: facet, grids and rel plot

# Importing Seaborn and setting styles
Behind the scenes, seaborn uses matplotlib to draw its plots. Through the application of the the default theme of seaborn, it uses the matplotlib rcParam system and will affect all matplotlib plots look, even if you don't make them with seaborn

```python
import seaborn as sns
sns.set_theme()
```
